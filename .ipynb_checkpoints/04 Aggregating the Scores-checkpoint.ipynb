```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd
import seaborn as sns
%matplotlib inline
```

## Read in the address units

```python
units=pd.read_csv('data/dc_wscores.csv')
units.rename(columns={'CLUSTER_':'NAME'}, inplace=True)
```

```python
# Examine the dataset.
print(units.shape)
print(units.isnull().sum())
units.head()
```

```python
units['hood']=units['ASSESSMENT_NBHD'].apply(lambda x: x.upper())
units.head()
```

```python
print(units['ASSESSMENT_NBHD'].nunique()) # Should be 71
print(units['NAME'].nunique()) # Should be 36
```

## Merge in the cluster names

```python
clusters=pd.read_csv('data/Clusters/Neighborhood_Clusters.csv')
clusters.head()
```

```python
full=pd.merge(units, clusters, on='NAME', how='left')
print(full.shape)
full.head()
```

## Aggregate scores by cluster

```python
full.columns
```

```python
scored_clusters=full.groupby('NAME').mean().reset_index().drop(['LATITUDE', 'LONGITUDE', 'ZIPCODE', 
                                      'OBJECTID', 'Shape_Length','Shape_Area', ], axis=1)
scored_clusters.head()
```

```python
# Merge back in the neigbhorhood names
names=clusters[['NBH_NAMES', 'NAME']]
print(len(scored_clusters))
scored_clusters=pd.merge(names, scored_clusters, on='NAME');
print(len(scored_clusters))
```

```python
for column in ['Units', 'Residential', 'mobilityscore', 'walkscore']:
    scored_clusters[column]=scored_clusters[column].apply(lambda x: round(x,1))
```

```python
print(scored_clusters.shape)
scored_clusters.head()
```

## Best & worst scores

```python
scored_clusters.rename(columns={'NBH_NAMES':'Neighborhood'}, inplace=True)
```

```python
# Top five neighborhoods for mobility
pd.set_option('display.max_colwidth', 200)
scored_clusters[['Neighborhood', 'mobilityscore']].sort_values(['mobilityscore'], ascending=False).head()
```

```python
# WORST five neighborhoods for mobility
pd.set_option('display.max_colwidth', 200)
scored_clusters[['Neighborhood', 'mobilityscore']].sort_values(['mobilityscore'], ascending=False).tail()
```

```python
# Top five neighborhoods for walking
scored_clusters[['Neighborhood', 'walkscore']].sort_values(['walkscore'], ascending=False).head()
```

```python
# WORST five neighborhoods for walking
pd.set_option('display.max_colwidth', 200)
scored_clusters[['Neighborhood', 'walkscore']].sort_values(['walkscore'], ascending=False).tail()
```

## Visualizing with a Shapefile

```python
shapes = gpd.read_file('data/Clusters/Neighborhood_Clusters.shp')
shapes.shape
```

```python
mymap=shapes.plot(figsize=(8,8))
mymap
```

```python
scored_shapes=pd.merge(scored_clusters, shapes, on='NAME', how='left')
scored_shapes.head(1)
```

```python
scored_shapes.columns
```

```python
df = gpd.GeoDataFrame(scored_shapes, geometry='geometry')
```

## Here come the maps

```python
# All DC neighborhoods
df.plot(figsize=(8,8));
```

```python
# Mobility Score
df.plot(column='mobilityscore', cmap='Reds', alpha=1, categorical=False, legend=True, ax=None, figsize=(12,8));
```

```python
# Walk Score
df.plot(column='walkscore', cmap='Greens', alpha=1, categorical=False, legend=True, ax=None, figsize=(12,8));
```

```python
# Residential
df.plot(column='Residential', cmap='Blues', alpha=1, categorical=False, legend=True, ax=None, figsize=(12,8));
```

```python
# Density of address units
df.plot(column='Units', cmap='Greys', alpha=1, categorical=False, legend=True, ax=None, figsize=(12,8));
```
